# **Retrival Augmented Generation (RAG) with a scientific paper**

## 1. Create Database
For image, I leverage LLM to generate its description. Then, create a dataframe with its source and its description for retrieval. Langchain in this part is used to preprocess the text: loading, chunking and embedding. The embedding model is text-embedding-3-small which provides 1536 dimension and suitable with the LLM used for retrieval.

Finally, the embedded text will be upserted to ChromaDB, one of the free vector database for POC.

## 2. Retrieval
For the retrieval, we use semantic search to find to relevant information

## 3. Interface
For interface, I use Gradio which allows me to customize the inference for the chatbot

### 3.1 Text Chatbot
The text chatbot could only receive text as input and output. To use it, please run ```python3 rag_chatbot.py```


### 3.2 Multimodel Chatbot
The multimodel Chatbot could only receive text as input but return both text answer and relevant figure.

To use it, please run rag_multimodel_chatbot.py
